```python
from datetime import date
import matplotlib.pyplot as plt
from deribit import MarketData
from deribit.market_data.options import premium_vs_expiry, premium_vs_strike
```

```python
async with MarketData() as client:
  X, Y = await premium_vs_strike(client, expiry=date(2025, 3, 20), currency='ETH', type='call')
  index = await client.get_index_price('eth_usd')
plt.plot(X, Y, label='Premium')
plt.axvline(index.index_price, label='Index Price', color='red')
plt.xlabel('Strike Price')
plt.ylabel('Premium')
plt.legend()
```

```python
async with client:
  X, Y = await premium_vs_expiry(client, strike=2000, currency='ETH', type='put')
plt.plot(X, Y, label='Premium')
plt.xlabel('Time to Expiry [days]')
plt.ylabel('Premium')
plt.legend()
```
